SQLAlchemy is NOT a database driver itself - it's a toolkit that sits on top of database drivers.

SQLAlchemy sees postgresql:// and thinks "I need a PostgreSQL driver to actually connect to this database." It then tries to import psycopg2 (the default PostgreSQL driver) behind the scenes.

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

getting DDL schema

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

now creating the table in postgres

```python
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

we are ingesting the data in chunks not all together so we create a new dataframe that reads the data from same url but in the chunks of 100000.

```python
df_iter = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)
```

we just iterated over these chunks and now we ingest the data into our postgresdatabase

to see progress:

```python
!uv add tqdm
```

```python
from tqdm.auto import tqdm
```

```python
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```
